```python
def writeFloatListToFile(floatList, fileName):
    f = open(fileName,'w')
    for flt in floatList:
        f.write('%f \n'%flt)
    f.close()
```

```python
myfavoriteFloats = [34.6,38.2,100.1]
outName = 'floatFile'
writeFloatListToFile(myfavoriteFloats,outName)
```

```python
def sumOfFloats(floatList):
    floatSum = 0
    for element in floatList:
        floatSum = floatSum + element
        
    return floatSum
```

```python
floatSumOut = sumOfFloats(myfavoriteFloats)
print(floatSumOut)
```
